```python
import mne
from ModelFamily import SCCNet, HSCNN, SCCTransformer, EEGNet
import torch
import numpy as np
import matplotlib.pyplot as plt 
```

```python
model = SCCNet()
model.load_state_dict(torch.load('SCCNet/SCCNet_S1_Iteration_0'), False)
for name, param in model.named_parameters():
    if name == 'spatial.0.weight':
        spatial_weight = param
```

```python
mont1020 = mne.channels.make_standard_montage('standard_1020')
kept_channels = [
    "Fp1",
    "Fp2",
    "F3",
    "F4",
    "C3",
    "C4",
    "P3",
    "P4",
    "O1",
    "O2",
    "F7",
    "F8",
    "T7",
    "T8",
    "P7",
    "P8",
    "Fz",
    "Cz",
    "Pz",
    "A1",
    "A2",
    "T1",
    "T2",
    "C6"
]
```

```python
ind = [i for (i, channel) in enumerate(mont1020.ch_names) if channel in kept_channels]
mont1020_new = mont1020.copy()
mont1020_new.ch_names = [mont1020.ch_names[x] for x in ind]
kept_channel_info = [mont1020.dig[x+3] for x in ind]
mont1020_new.dig = mont1020.dig[0:3]+kept_channel_info
n_channels = len(mont1020_new.ch_names)

fake_info = mne.create_info(ch_names=mont1020_new.ch_names, sfreq=250.,
                            ch_types='eeg')

rng = np.random.RandomState(0)
data = rng.normal(size=(n_channels, 1)) * 1e-6

fake_evoked = mne.EvokedArray(data, fake_info)
fake_evoked.set_montage(mont1020_new)
```

```python
figure, ax = plt.subplots(2, 2)
mne.viz.plot_topomap(spatial_weight[0,0,:,0].detach().numpy(), fake_evoked.info, axes=ax[0,0])
mne.viz.plot_topomap(spatial_weight[1,0,:,0].detach().numpy(), fake_evoked.info, axes=ax[0,1])
mne.viz.plot_topomap(spatial_weight[2,0,:,0].detach().numpy(), fake_evoked.info, axes=ax[1,0])
mne.viz.plot_topomap(spatial_weight[3,0,:,0].detach().numpy(), fake_evoked.info, axes=ax[1,1])
```
